# Extracting Questions Database from Ti-Place

##Obtaining list of all Object IDs

```
import xml.etree.cElementTree as et
import pandas as pd
import urllib.request as urllib2

obj_tree = et.ElementTree(file=urllib2.urlopen('http://api.seescp.com/webservice/xml/taxonomy/list.tsty?svcid=tiplace'))
obj_root = obj_tree.getroot()

obj_data = []
for el in obj_tree.iterfind('.//*'):
    for i in el.iterfind('*'):
        obj_data.append(dict(i.items()))

obj_df = pd.DataFrame(obj_data)
obj_df
```

##Obtaining questions for one Object ID as a table - testing

```
q_tree = et.ElementTree(file=urllib2.urlopen('http://api.seescp.com/webservice/tplace/read.tsty?svcid=tiplace&objId=61334'))
q_root = q_tree.getroot()

tags=[]
texts=[]
attribs=[]

for elem in q_root.iter():
  tags.append(elem.tag)
  texts.append(elem.text)
  attribs.append(elem.attrib)
  
q_df = pd.DataFrame({"TAG": tags, "TEXT": texts, "ATTRIBUTE": attribs})
q_df = q_df[q_df.TAG != 'activityElement']
q_df['ATTRIBUTE'] =  q_df['ATTRIBUTE'].map(lambda x: str(x)[:-2])
q_df['ATTRIBUTE'] =  q_df['ATTRIBUTE'].map(lambda x: str(x)[8:])
q_df['ATTRIBUTE']=q_df.ATTRIBUTE.str.strip()
q_df['ATTRIBUTE']=pd.to_numeric(q_df.ATTRIBUTE)
q_df['ATTRIBUTE_RANK'] = q_df['ATTRIBUTE'].rank()
q_df = q_df.fillna('')
q_df['ATTRIBUTE_RANK'] =  q_df['ATTRIBUTE_RANK'].map(lambda x: str(x)[:-2])
q_df.ATTRIBUTE = q_df.ATTRIBUTE_RANK.astype(str)
q_df['ATTRIBUTE_TAG']=q_df.apply(lambda x:'%s%s' % (x['TAG'],x['ATTRIBUTE_RANK']),axis=1)
q_df=q_df.drop(['ATTRIBUTE','TAG','ATTRIBUTE_RANK'],axis=1)
q_df_transposed=q_df.set_index('ATTRIBUTE_TAG').T

q_df_transposed.dtypes
```

##Obtaining questions for one Object ID as a table - testing

```
small_obj = obj_df.head(10)
master_q_df = pd.DataFrame(columns=['testItemList','code','totalcount','testItem','objid','title','score','weight','taxonomy','node1','node2','node3','node4','node5',
                                    'node6','previewImage','meta','learningStrategy','itemStrategy','itemDifficulty','itemdifficulty9','itemdifficulty5','itemType',
                                    'isOnline','istexerror','activityElements']) 

for item in small_obj['id']:
  url='http://api.seescp.com/webservice/tplace/read.tsty?svcid=tiplace&objId=' + item
  q_tree = et.ElementTree(file=urllib2.urlopen(url))
  q_root = q_tree.getroot()
  
  if len(q_root) > 2:
    tags=[]
    texts=[]
    attribs=[]
    
    for elem in q_root.iter():
      tags.append(elem.tag)
      texts.append(elem.text)
      attribs.append(elem.attrib)
      q_df = pd.DataFrame({"TAG": tags, "TEXT": texts, "ATTRIBUTE": attribs})
      q_df = q_df[q_df.TAG != 'activityElement']
      q_df['ATTRIBUTE'] =  q_df['ATTRIBUTE'].map(lambda x: str(x)[:-2])
      q_df['ATTRIBUTE'] =  q_df['ATTRIBUTE'].map(lambda x: str(x)[8:])
      q_df['ATTRIBUTE']=q_df.ATTRIBUTE.str.strip()
      q_df['ATTRIBUTE']=pd.to_numeric(q_df.ATTRIBUTE)
      q_df['ATTRIBUTE_RANK'] = q_df['ATTRIBUTE'].rank()
      q_df = q_df.fillna('')
      q_df['ATTRIBUTE_RANK'] =  q_df['ATTRIBUTE_RANK'].map(lambda x: str(x)[:-2])
      q_df.ATTRIBUTE = q_df.ATTRIBUTE_RANK.astype(str)
      q_df['ATTRIBUTE_TAG']=q_df.apply(lambda x:'%s%s' % (x['TAG'],x['ATTRIBUTE_RANK']),axis=1)
      q_df=q_df.drop(['ATTRIBUTE','TAG','ATTRIBUTE_RANK'],axis=1)
      q_df_transposed=q_df.set_index('ATTRIBUTE_TAG').T
      master_q_df=master_q_df.merge(q_df_transposed, on=None, how='outer')
```
