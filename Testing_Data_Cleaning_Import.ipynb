```
# Renaming columns names
raw2=raw1.rename(columns={'시험지':'TEST_ID',
                          '문항':'QUESTION_ID',
                          '학생코드':'STUDENT_ID',
                          '이름':'NAME',
                          '정오':'RESPONSE',
                          'Total Time':'TOTAL_TIME',
                          'Initial Time':'INITIAL_TIME',
                          'Solving Time':'SOLVING_TIME',
                          'Delay Time(1s)':'TOTAL_DELAY_1S',
                          'Of Stroke':'STROKE_COUNT',
                          '1s':'DELAY_1S',
                          '2s':'DELAY_2S',
                          '3s':'DELAY_3S',
                          '4s':'DELAY_4S',
                          '5s':'DELAY_5S',
                          '10s':'DELAY_10S',
                          '20s':'DELAY_20S',
                          '30s':'DELAY_30S',
                          'Total Ink Length':'TOTAL_INK_LENGTH',
                          'Avg Speed':'AVERAGE_SPEED',
                          'Inititation Speed':'INITIATION_SPEED',
                          'Termination Speed':'TERMINATION_SPEED',
                          '영역값':'REWORK',
                         })
```

```
raw2.loc[raw2['RESPONSE'] == 'X', 'RESPONSE'] = '0'
raw2.loc[raw2['RESPONSE'] == 'O', 'RESPONSE'] = '1'
```

```
raw2['TEST_ID'] = raw2['TEST_ID'].apply(str)
raw2['QUESTION_ID'] = raw2['QUESTION_ID'].apply(str)
raw2['STUDENT_ID'] = raw2['STUDENT_ID'].apply(str)
raw2['RESPONSE'] = pd.to_numeric(raw2.RESPONSE)
```

```
#Import the question mapping dataset

map1=pd.read_excel('/content/drive/My Drive/cai_data/question_mapping_nov_2018.xlsx',index_col=None)
map1.head()
```

```
# Renaming columns names
map2=map1.rename(columns={'문항번호':'QUESTION_ID',
                          '문항제목':'QUESTION_TITLE',
                          '구성영역':'CLASS_LEVEL',
                          '학년':'CLASS_GRADE',
                          '교과목':'SUBJECT',
                          '내용영역':'QUESTION_CATEGORY',
                          '대영역':'QUESTION_MAIN_CATEGORY',
                          '난이도':'QUESTION_DIFFICULTY_PRESET',
                          '풀이시간':'SOLVING_TIME_PRESET',
                          '배점':'QUESTION_SCORE',
                          '정답률':'CORRECT_RATE_PRESET',
                          '문항푼횟수':'COUNT_USED_PRESET'
                         })

map2['QUESTION_ID'] = map2['QUESTION_ID'].apply(str)
```

```
map2.head()
```

```
map2['QUESTION_CATEGORY'].unique().tolist()
```

```
map2.loc[map2['QUESTION_CATEGORY'] == '수와 연산', 'QUESTION_CATEGORY'] = 'Numbers and Arithmetic'
map2.loc[map2['QUESTION_CATEGORY'] == '도형', 'QUESTION_CATEGORY'] = 'Shapes'
map2.loc[map2['QUESTION_CATEGORY'] == '확률과 통계', 'QUESTION_CATEGORY'] = 'Probability and Statistics'
map2.loc[map2['QUESTION_CATEGORY'] == '규칙', 'QUESTION_CATEGORY'] = 'Sequences'
map2.loc[map2['QUESTION_CATEGORY'] == '측정', 'QUESTION_CATEGORY'] = 'Measurements'
map2.loc[map2['QUESTION_CATEGORY'] == '문자와 식', 'QUESTION_CATEGORY'] = 'Equations'
map2.loc[map2['QUESTION_CATEGORY'] == '함수', 'QUESTION_CATEGORY'] = 'Functions'
map2.loc[map2['QUESTION_CATEGORY'] == '기하', 'QUESTION_CATEGORY'] = 'Geometry'
map2.loc[map2['QUESTION_CATEGORY'] == '도형의방정식', 'QUESTION_CATEGORY'] = 'Geometric Equations'
map2.loc[map2['QUESTION_CATEGORY'] == '화학', 'QUESTION_CATEGORY'] = 'Chemistry'
map2.loc[map2['QUESTION_CATEGORY'] == '생물', 'QUESTION_CATEGORY'] = 'Biology'
map2.loc[map2['QUESTION_CATEGORY'] == '지구과학', 'QUESTION_CATEGORY'] = 'Geography'
map2.loc[map2['QUESTION_CATEGORY'] == 'Number', 'QUESTION_CATEGORY'] = 'Numbers and Arithmetic'
map2.loc[map2['QUESTION_CATEGORY'] == 'Measure', 'QUESTION_CATEGORY'] = 'Measurements'
map2.loc[map2['QUESTION_CATEGORY'] == 'Numbers', 'QUESTION_CATEGORY'] = 'Numbers and Arithmetic'
map2.loc[map2['QUESTION_CATEGORY'] == '해석', 'QUESTION_CATEGORY'] = 'Analysis'
map2.loc[map2['QUESTION_CATEGORY'] == 'KMO 1차', 'QUESTION_CATEGORY'] = 'KMO'
map2.loc[map2['QUESTION_CATEGORY'] == '대수', 'QUESTION_CATEGORY'] = 'Algebra'
map2.loc[map2['QUESTION_CATEGORY'] == '집합과명제', 'QUESTION_CATEGORY'] = 'Set Theory'
map2.loc[map2['QUESTION_CATEGORY'] == '경시', 'QUESTION_CATEGORY'] = 'Olympiad'
map2.loc[map2['QUESTION_CATEGORY'] == '수와연산', 'QUESTION_CATEGORY'] = 'Numbers and Arithmetic'
map2.loc[map2['QUESTION_CATEGORY'] == '물리', 'QUESTION_CATEGORY'] = 'Physics'
map2['QUESTION_CATEGORY']=map2[['QUESTION_CATEGORY']].fillna(value='Not Available')
```

```
map2['QUESTION_DIFFICULTY_PRESET'].unique().tolist()
```

```
map2.loc[map2['QUESTION_DIFFICULTY_PRESET'] == '하', 'QUESTION_DIFFICULTY_PRESET'] = 'LOW'
map2.loc[map2['QUESTION_DIFFICULTY_PRESET'] == '중', 'QUESTION_DIFFICULTY_PRESET'] = 'MED'
map2.loc[map2['QUESTION_DIFFICULTY_PRESET'] == '상', 'QUESTION_DIFFICULTY_PRESET'] = 'HIGH'
map2['QUESTION_DIFFICULTY_PRESET']=map2[['QUESTION_DIFFICULTY_PRESET']].fillna(value='Not Available')
```

```
raw3=pd.merge(raw2, map2[['QUESTION_ID','QUESTION_CATEGORY','QUESTION_DIFFICULTY_PRESET','SOLVING_TIME_PRESET','QUESTION_SCORE','CORRECT_RATE_PRESET','COUNT_USED_PRESET']], on='QUESTION_ID', how='left')
```
